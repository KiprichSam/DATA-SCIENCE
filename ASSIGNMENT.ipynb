# **2. Compare Anaconda with pip/virtualenv for managing environments and packages?**

## **4. What are the key differences between Jupyter Notebook and JupyterLab?** *italicised text*

1. Jupyter Notebook has a simple, linear interface where each notebook opens in its own new browser tab focusing on single notebook document at a time whereas JupyterLab offers a multi-document interface (MDI) within a single browser tab providing a flexible, integrated workspace where you can open multiple notebooks, code consoles, text editors, terminals, and even display output views.

2. JupyterLab is slightly heavier and more resource-intensive due to its more complex interface and integrated features, especially with many tabs or panels open as compared to Jupyter Notebook.

3. Jupyter Notebook is mainly limited to working with .ipynb files, whereas JupyterLab supports a variety of file types such as Python scripts, Markdown, CSVs, and JSON files in one integrated workspace.

4. Jupyter Notebook is lightweight and beginner-friendly with a minimal interface, while JupyterLab offers a more powerful, IDE-like environment that is customizable and better suited for advanced workflows.

## **5. Explain how Git and GitHub work together. What is a pull request and when is it used?**

Git is a version control system that is installed and running locally on your machine primarily tracking file changes, managing different versions of a project, and allowing multiple people to work on the same project without overwriting each other's work.

GitHub is a web-based hosting service for Git repositories. While Git handles the mechanics of version control on your local machine, GitHub provides a centralized, cloud-based platform where you can store your Git repositories.

GitHub adds a layer of collaborative tools, a graphical user interface (GUI), and social features on top of Git, making it much easier for teams to work together.

A pull request is a formal proposal to merge a set of changes from one branch into another branch in a Git repository.
Pull requests are used in almost any collaborative software development or data science project that uses Git and a hosting service like GitHub.
